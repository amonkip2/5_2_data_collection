The data presented has been sourced from kaggele site https://www.kaggle.com/wendykan/lending-club-loan-data. The data represents loans issued through the 2007-2015, including the current loan status (Current, Late, Fully Paid, etc.) and latest payment information. The file containing loan data through the "present" contains complete loan data for all loans issued through the previous completed calendar quarter. Additional features include credit scores, number of finance inquiries, address including zip codes, and state, and collections among others. The file obtained is a compressed csv file (loans.csv.tar.gz) which contains over 890 thousand observations and 75 variables. A data dictionary providing description of columns is also i.e. LCDataDictionary.xlsx is included.
